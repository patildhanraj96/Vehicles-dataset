# Data set name- Vehicles data
##            About features of vehicle dat set :cars
1) Name / Brand - Name of the cars

2) Year - Year of the car when it was bought

3) selling_price - Price at which the car is being 

4) km_driven - Number of Kilometres the car is driven

5) fuel - Fuel type of car
 
6) seller_type - tells if a seller is individual or a dealer

7) transmission - Gear transmission fo the car

8) owner - Number of previous owners of the car

9) mileage_kmpl - mileage in km per litre of fuel

10) engine_CC - engine size in cubic cm

11) max_power_bhp - max break horse power of car

12) seats - no of seats available in car 

13) torque_Nm - torque genrated by car

14) RPM - revolution per minute (speed) of car
